# Radioactive Decay Simulation

## Overview
This notebook simulates radioactive decay using the exponential distribution, which describes the time intervals between random decay events in nuclear physics.

## Key Concepts

### Exponential Distribution
The time between consecutive decay events follows an exponential distribution. This is a fundamental principle in nuclear physics where:
- **λ (decay constant)**: Determines the decay rate
- **Mean decay time**: τ = 1/λ
- The probability of decay is constant over time (memoryless property)

### Parameters Used
- **N = 1000**: Number of particles simulated
- **λ = 0.1 s⁻¹**: Decay constant (per second)
- **Expected mean decay time**: 1/0.1 = 10 seconds

## Implementation Details

`np.random.exponential()` generates random samples from an exponential distribution where:
- `scale = 1/λ` sets the mean value
- Each sample represents the time until one particle decays
- Results approximate the theoretical distribution with 1000 samples

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 1000  # Number of particles
decay_constant = 0.1  # Lambda (per second)

# Generate decay times using exponential distribution
# Physics: Time between events follows exponential distribution
decay_times = np.random.exponential(scale=1/decay_constant, size=N)

print(f"Generated {N} decay times")
print(f"First 10 decay times: {decay_times[:10]}")
print(f"Mean decay time: {np.mean(decay_times):.10f} seconds")
print(f"Theoretical mean: {1/decay_constant:.10f} seconds")
```

```python
# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(decay_times, bins=50, density=True, alpha=0.7, 
         color='blue', edgecolor='black', label='Simulated Data')

# Overlay theoretical curve
t = np.linspace(0, max(decay_times), N)
theoretical = decay_constant * np.exp(-decay_constant * t)
plt.plot(t, theoretical, 'r-', linewidth=2, label='Theoretical Curve')

plt.xlabel('Decay Time (seconds)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.title('Monte Carlo Simulation of Radioactive Decay', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
